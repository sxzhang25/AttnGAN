```python
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## Generating images + attention

```python
filename = '../data/birds/example_captions.txt'
lines = ['this is a blue bird with black tip wings and a black tail']
num_lines = 50

file = open(filename, 'w')
for line in lines:
    for i in range(num_lines):
        file.writelines(line + '\n')
file.close()
```

```python
dirname = '../models/bird_AttnGAN2/example_captions'
images = []
for filename in os.listdir(dirname):
    if filename.endswith("_g2.png"): 
        images.append(filename)
    else:
        continue
images.sort()
```

```python
cols = 5
rows = np.ceil(len(images) / cols)
fig = plt.figure(figsize=(2 * rows, 5 * cols))
for i,image in enumerate(images):
    ax = fig.add_subplot(rows, cols, i+1)
    ax.imshow(cv2.imread(os.path.join(dirname, images[i]))[...,::-1])
plt.show()
```
